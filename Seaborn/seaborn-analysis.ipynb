# Introduction to Seaborn

Seaborn is a data visualization library in Python. It is used to produce statistical graphics in Python. It is built on top of Matplotlib. It also supports NumPy and Pandas data structures. It also supports statistical units from SciPy.

Seaborn has various types of plots which can satisfy various types of data visualization requirements. The plots can be categorized into seven categories â€“ relational plots, categorical plots, distribution plots, regression plots, matrix plots and multi-plot grids.

# Seaborn Color Palette-set()

```python
current_palette2 = sns.color_palette()
sns.palplot(sns.color_palette("Greens"))
```

```python
sns.palplot(current_palette2)
```

# Plotting Univariate Distribution with displot()

The most important thing to do while analysing the data is to understand its distribution. Seaborn helps us to understand the univariate distribution of data.

The distplot() function provides a quick look at univariate distribution.

```python
sns.displot(iris['sepal_length']) 
```

# Plotting Bivariate Distribution with jointplot()

Bivariate Distribution is used to determine the relation between two variables. This mainly deals with relationship between two variables and how one variable is behaving with respect to the other.

The best way to analyze Bivariate Distribution in seaborn is by using the jointplot() function.

```python
sns.jointplot(x="species", y="sepal_length", data=iris)
```

# Seaborn - Scatter plot

A scatter plot can be used to demonstrate relationship between two variables x and y.

```python
sns.scatterplot(x="petal_length", y="petal_width", data=iris,hue="species")
```

# Visualizing pairwise relationship with pairplot()

Some datasets contain many variables. In such cases, the relationship between each and every variable should be analyzed. So, we need to plot pairwise relationships in a dataset.

To plot multiple pairwise bivariate distributions, we can use pairplot() function. 

```python
sns.pairplot(iris,hue='species')
```

# Plotting categorical data

A stripplot() is used to draw a scatterplot where one variable is categorical. It represents the data in sorted order along any one of the axis.

```python
sns.stripplot(y="sepal_length", x="species", data=iris,hue="species")
```

 swarmplot() can be used to draw categorical scatterplot with non-overlapping points.

```python
sns.swarmplot(y="sepal_length", x="species", data=iris,hue="species")
```

# Distribution of Observations

## Box Plot

Box plots usually have vertical lines extending from the boxes which are termed as whiskers. These whiskers indicate variability outside the upper and lower quartiles, hence Box Plots are also termed as box-and-whisker plot.

```python
sns.boxplot(x=iris["sepal_length"])
```

# Statistical estimation with Seaborn

## Bar Plot

A barplot shows the relation between a categorical variable and a continuous variable. The data is represented in rectangular bars where the length the bar represents the proportion of the data in that category.

```python
sns.barplot(x="species",y="sepal_length", data=iris)
```

## Point Plot

```python
sns.pointplot(y="sepal_length", x="species", data=iris)
```

# Linear relationships with Seaborn

## lmplot

It is intended as a convenient interface to fit regression models across conditional subsets of a dataset.

```python
sns.lmplot(x="petal_length", y="petal_width", data=iris,hue="species")
```

# Matrix plots with Seaborn

## Heat map

Seaborn heatmap() function plot rectangular data as a color-encoded matrix

```python
sns.heatmap(iris[['sepal_length','sepal_width','petal_length','petal_width']].corr(),
           annot=True,linewidth=2,linecolor='yellow')
```
